```python
import pandas as pd
import bs4
from bs4 import BeautifulSoup
import requests
import json
```

```python
#urls = BeautifulSoup(open("https://humanist.kdl.kcl.ac.uk/Archives/Converted_Text/"),features="html.parser")
response = requests.get("https://humanist.kdl.kcl.ac.uk/Archives/Converted_Text/")
soup = BeautifulSoup(response.text, features="html.parser")
results = soup.find_all('a')
#results_dict = {}
titles = []
text_list = []

'''for result in results:
    result = result.get_text()
    #print(result)
    results_dict['Volume'].append(result)'''

links = soup.find_all('a')
for url in links:
    #link_text = url.get_text()
    url = url.get('href')
    titles.append(url)
    #print(link, link_text)

del titles[-1]
del titles[0]    

correct_url = [i.replace('humanist','https://humanist.kdl.kcl.ac.uk/Archives/Converted_Text/humanist') for i in titles]
correct_url

# I failed to extract only #text by using .get, so received some ideas from Jackey's
sub = soup.text.split()[6:-4]
date = sub[1::4]
time = sub[2::4]
volumes = sub[3::4]
```

```python
for i in range(len(titles)):
    combind_dict = {'title':titles[i], 'URL':correct_url[i],
                    'Date':date[i], 'Time':time[i], 'Volume':volumes[i]}
    text_list.append(combind_dict)
text_list
```

```python
humanist_vols = pd.DataFrame.from_dict(text_list, orient='columns')
# Rearrange order
humanist_vols = humanist_vols[['title','URL','Date','Time','Volume']]
humanist_vols
```
